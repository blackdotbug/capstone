```python
import psycopg2
from sqlalchemy import create_engine
import config
import pandas as pd

DATABASE_URL = config.postgresURI
```

```python
engine = create_engine(DATABASE_URL)
```

```python
sexism_data = pd.read_csv("../data/sexism/sexism_data.csv")
sexism_data
```

```python
sexism_data = sexism_data.sample(n=10000, random_state=919)
sexism_data
```

```python
sexism_data.to_sql('sexism_data', con = engine, if_exists='replace')
```
